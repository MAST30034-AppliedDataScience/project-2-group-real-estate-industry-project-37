```python
import matplotlib.pyplot as plt

# Defining the regions and their growth rates
regions = ['South Eastern Melbourne', 'Outer Eastern Melbourne', 'Southern Melbourne', 
           'Inner Eastern Melbourne', 'North Western Melbourne', 'North Eastern Melbourne', 'Inner East Melbourne']
growth_rates = [1.055, 1.011, 0.830, 0.779, 0.751, 0.751, 0.570]

# Creating a bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(regions, growth_rates, color='skyblue')
plt.xlabel('Growth Rate (%)')
plt.title('Growth Rates of Melbourne Regions')

# Adding data labels
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, 
             f'{bar.get_width():.3f}%', 
             va='center', ha='left', color='black', fontsize=10)

plt.gca().invert_yaxis()  # Reverse the order to display the highest growth rate at the top
plt.show()
```
