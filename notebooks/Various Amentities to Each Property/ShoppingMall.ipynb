# Use API to Find Shopping Mall Specific Locations

```python
import googlemaps
import time

api_key = 'Your Keys API'
gmaps = googlemaps.Client(key=api_key)

victoria_centers = [
    (-37.8136, 144.9631),
    (-38.1499, 144.3617),
    (-37.5632, 143.8496),
    (-36.7588, 144.2802),
    (-38.3402, 145.1235),
    (-37.7069, 145.0000),
    (-37.8303, 144.9592),
    (-37.5502, 144.3003),
    (-38.3667, 144.9680), 
    (-36.7805, 145.5692),
    (-37.7321, 145.3451),
    (-37.0738, 145.9416),
    (-36.1223, 144.7461),
    (-38.1441, 146.4659), 
    (-36.6733, 142.1998), 
    (-37.5836, 145.2663),
    (-38.3830, 142.4805),
    (-37.2205, 146.4311),
    (-38.6800, 146.0150), 
    (-37.0000, 144.6678)
]

# Use the Google Places API to search for shopping centers in multiple regions
shopping_malls = []
next_page_token = None

for center in victoria_centers:
    while True:
        result = gmaps.places_nearby(location=center, radius=50000, type='shopping_mall', page_token=next_page_token)

        shopping_malls.extend(result['results'])

        next_page_token = result.get('next_page_token')
        if not next_page_token:
            break
        time.sleep(2)

total_malls = len(shopping_malls)

for mall in shopping_malls:
    print(f"Name: {mall['name']}, Address: {mall.get('vicinity', 'No Location')}, Longitude: {mall['geometry']['location']['lng']}, Latitude: {mall['geometry']['location']['lat']}")

print(f"Total{total_malls}")
```

```python
import pandas as pd

data = []

# Walk through all the shopping centers, adding name, address, longitude, and latitude to the data list
for mall in shopping_malls:
    name = mall['name']
    address = mall.get('vicinity', 'Unknown Address')
    longitude = mall['geometry']['location']['lng']
    latitude = mall['geometry']['location']['lat']

    if address != 'Unknown Address':
        data.append([name, address, longitude, latitude])

# Create a DataFrame using pandas
df = pd.DataFrame(data, columns=['Name', 'Address', 'Longitude', 'Latitude'])

df.to_csv('../../ADS ASS2/Data/shopping_malls.csv', index=False, encoding='utf-8-sig')
```
